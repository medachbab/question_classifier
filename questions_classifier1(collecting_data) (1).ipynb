```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
url= 'https://www.signupgenius.com/groups/getting-to-know-you-questions.cfm'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
#print(soup)
```

```python
list_elements = soup.find_all('ol')
#list_elements
#liste pour stockage des questions
questions_collection = []
for result_lists in list_elements:
    list_questions=result_lists.find_all('li')
     #list_questions
    for question in list_questions:
        text = question.get_text(strip=True)
        #print(text)
        questions_collection.append([text, 1])
df = pd.DataFrame(questions_collection, columns=['question', 'personal'])
df.drop_duplicates(inplace=True)
df
```

```python
list_elements = soup.find_all('p')
#list_elements
#liste pour stockage des questions
questions_collection = []
for result_lists in list_elements:
    list_questions=result_lists.find_all('span')
    for question in list_questions:
        text = question.get_text(strip=True)
        if re.match(r'^\d', text):
            #print(text)
            questions_collection.append([text, 1])
df1 = pd.DataFrame(questions_collection, columns=['question', 'personal'])
df1.drop_duplicates(inplace=True)
df1
```

on doit eleminer les numereau au debut de chaque lignes ainsi les lignes vides:

```python
df1['question'] = df1['question'].replace(r'^\d+\.\s*', '', regex=True)
#df1.info()
df1
df1 = df1[df1['question'].str.strip() != '']
df1
```

```python
list_elements = soup3.find_all('ol')
#list_elements
#liste pour stockage des questions
questions_collection = []
for result_lists in list_elements:
    list_questions=result_lists.find_all('li')
    for question in list_questions:
        text = question.get_text(strip=True)
        #print(text)
        questions_collection.append([text, 0])
df3 = pd.DataFrame(questions_collection, columns=['question', 'personal'])
df3
```

```python
df3= df3[df3['question'].str.startswith('Question:')]
df3['question']= df3['question'].str.replace(r'^Question:\s*', '', regex=True)
df3
```
